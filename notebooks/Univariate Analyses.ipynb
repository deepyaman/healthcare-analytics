# Plot univariate distributions

```python
# https://stackoverflow.com/a/40148917/1093967
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.api.types import is_numeric_dtype
from pylab import rcParams

numeric_columns = [c for c in X.columns if is_numeric_dtype(X[c])]
width, height = rcParams["figure.figsize"]
fig, axs = plt.subplots(
    len(numeric_columns),
    2,
    sharey=False,
    figsize=(2 * width, len(numeric_columns) * height),
)

joined.boxplot(
    column=numeric_columns,
    by="tgt_health_camp_outcome_favorable",
    ax=axs[:, 0],
    layout=(len(numeric_columns), 1),
)
[ax.set_xlabel("") for ax in axs[:, 0].reshape(-1)]

for i, column in enumerate(numeric_columns):
    dropzero = column not in ["Income", "Education_Score", "Age", "Category3"]
    weights = 1 / y[X[column] > 0 if dropzero else X[column].notna()].groupby(
        "tgt_health_camp_outcome_favorable"
    ).tgt_health_camp_outcome_favorable.transform("count")
    sns.histplot(
        joined,
        x=column,
        hue="tgt_health_camp_outcome_favorable",
        weights=weights,
        element="step",
        legend=not i,  # Only show the legend for the first plot
        ax=axs[i, 1],
    )
    axs[i, 1].set_title(f"{column} (N={X[column].notna().sum()}, {dropzero=})")
    axs[i, 1].set_xlabel("")
    axs[i, 1].set_ylabel("")

plt.show()
```
